```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = io.imread('data/images/pos/abigotte_pos_001.jpg')
plt.imshow(image)
```

```python
# compute the whiteness of the image
# it can be defined as the inverse of the differences between the red, green and blue channels, multiplied by the sum of the channels
diffs = np.abs(image[:,:,0] - image[:,:,1]) + np.abs(image[:,:,0] - image[:,:,2]) + np.abs(image[:,:,1] - image[:,:,2])
# whiteness = np.ones_like(image[:,:,0]) * 255 - diffs / np.sum(image, axis=2)
intensity = np.mean(image, axis=2, dtype=np.uint8)
```
