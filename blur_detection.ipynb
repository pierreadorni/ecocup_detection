```python
import blur_detector
from skimage import io
from skimage.color import rgb2gray
from matplotlib import pyplot as plt
```

```python
# load image
image = rgb2gray(io.imread('data/images/pos/abigotte_pos_001.jpg'))
```

```python
# detect blur
map = blur_detector.detectBlur(image, downsampling_factor=4, num_scales=4, scale_start=2, num_iterations_RF_filter=3)
```

```python
# plot image and blur side to side
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Image')
ax[1].imshow(map)
ax[1].set_title('Blur map')
```

```python
# plot blur map over image
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.imshow(image, cmap='gray')
ax.imshow(map, alpha=0.5)
ax.set_title('Image with blur map')
```

```python
# whole process in a func
def plot_blur_map(image):
    map = blur_detector.detectBlur(image, downsampling_factor=4, num_scales=4, scale_start=2, num_iterations_RF_filter=3, sigma_s_RF_filter=15,sigma_r_RF_filter=0.01, show_progress=False)
    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    ax.imshow(image, cmap='gray')
    ax.imshow(map, alpha=0.5)
    ax.set_title('Image with blur map')
```

```python
plot_blur_map(rgb2gray(io.imread('data/images/pos/abigotte_pos_005.jpg')))
```
